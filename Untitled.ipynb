```python
songs_url = []

for rank in ranks_url:    
    res = requests.get(rank, headers=headers)
    
    if res.status_code == 200:
        songs_tag = BeautifulSoup(res.text,"html")

        songs = songs_tag.find("div",class_="pc_temp_songlist").find("ul").find_all("li")
        
        for song in songs:
            sn = song.find("a")
            songs_url.append(sn.get("href").strip())

print(songs_url)
```
